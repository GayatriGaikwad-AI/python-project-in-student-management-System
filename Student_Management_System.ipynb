## PROJECT IN PYTHON

```python
students=[]
courses=set()
college=("ABC College","pune")
print("Project Started")
```

## Add Student Function

```python
def add_student():
    sid=int(input("Enter Student ID:"))
    name=input("Enter Name:")
    course=input("Enter Course:")

    student={
        "id":sid,
        "Name":name,
        "Course":course
    }
    students.append(student)
    courses.add(course)

    print("student Added Successfully")
```

```python
add_student()
```

```python
print(students)
```

```python
def view_students():
    for student in students:
        print(student)
```

```python
view_students()
```

```python
def search_student():
    sid=int(input("Enter student ID:"))
    for student in students:
        if student["id"]==sid:
            print(student)
            return
    print("student not found")
```

```python
search_student()
```

```python
def delete_student():
    sid=int(input("Enter Student ID:"))
    for student in students:
        if student["id"]==sid:
            students.remove(student)
            print("student Deleted successfully")
            return
    print("student not found")
```

```python
print("Before Delete:",students)
delete_student()
print("After Delete:",students)
```

```python
def view_students():
    for student in students:
        print(student)
```

```python
view_students()
```
